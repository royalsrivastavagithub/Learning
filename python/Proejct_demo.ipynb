Importing libraries required 

```python
import yfinance as yf
import seaborn as sns
import matplotlib.pyplot as plt
```

Downloading Requiard data 

```python
reliance = yf.download("reliance.ns", start="1996-01-01", end="2023-03-13")
```

Visualization: 

```python
sns.lineplot(data=reliance["Close"], label="Close")
sns.lineplot(data=reliance["Open"], label="Open")
sns.lineplot(data=reliance["High"], label="High")
sns.lineplot(data=reliance["Low"], label="Low")
plt.legend()
plt.show()
```

```python
sns.scatterplot(data=reliance, x="Open", y="Close")
plt.xlabel("Open")
plt.ylabel("Close")
plt.show()
```

```python
sns.histplot(data=reliance["Volume"])
plt.xlabel("Volume")
plt.ylabel("Frequency")
plt.show()
```
